# **Clasificación de imágenes de quemaduras de segundo y tercer grado**

## Máquina de Soporte Vectorial

```python
import os
import cv2
import joblib
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
```

```python
img = cv2.imread('quemaduras/segundo_grado/23_grado_2.jpg')

imgs_bgr = (img.copy(), img.copy(), img.copy())

#Apagar los otros dos canales para cada imagen
for i in range(3):
    complemento = [k for k in range(3) if k != i]
    imgs_bgr[i][:, :, complemento] = 0

fig, ax = plt.subplots(2, 2, figsize=(6, 8))

ax[0, 0].imshow(imgs_bgr[0])
ax[0, 0].set_title("Red")

ax[0, 1].imshow(imgs_bgr[1])
ax[0, 1].set_title("Green")

ax[1, 0].imshow(imgs_bgr[2])
ax[1, 0].set_title("Blue")

ax[1, 1].imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
ax[1, 1].set_title("RGB")

plt.show()
```

## Blue

```python
width = 540
height = 960
ruta_carpeta = 'quemaduras/'

X = []
y = []

for i in os.listdir(ruta_carpeta):
    for j in os.listdir(ruta_carpeta + i):
        img = cv2.imread(ruta_carpeta + i + '/' + j)
        resized_img = cv2.resize(img, (width, height))
        blue = resized_img[:, :, 0]
        
        X.append(blue.flatten())

        if i == 'segundo_grado':
            y.append(0) 
        else:
            y.append(1)

X = np.array(X)
y = np.array(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)

svm = SVC(kernel='rbf')
svm.fit(X_train, y_train)

#Train
y_pred_train = svm.predict(X_train)

CM_train = confusion_matrix(y_train, y_pred_train)
accuracy_train = accuracy_score(y_train, y_pred_train)

print("Matriz de confusión train:\n", CM_train)
print("Accuracy:", accuracy_train)

#Test
y_pred = svm.predict(X_test)

CM = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print("\nMatriz de confusión test:\n", CM)
print("Accuracy:", accuracy)
print("Precision:", precision)
print("Recall:", recall)
print("F1 Score:", f1)
```
